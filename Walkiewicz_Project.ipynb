## *Web scraping*

Web scraping is the practice of collecting information from websites. 
In reality, online scraping may be equivalent to a user manually copying information from a webpage. This procedure, on the other hand, is totally automated and provides structured web data from a large number of pages. Scraping entails retrieving (downloading) and extracting data from a web page.

## *Natural Language Toolkit - NLTK*

The Natural Language Toolkit (NLTK) is another major tool used in the project. 
“NLTK is a premier framework for constructing Python programs to work with human language data,” according to the project's official website. It is used to prepare text preprocessing, which is required before any text analysis can be performed. The user can do the following with NTLK:
* develop a tokenization and tagging system
* determine the existence of named entities
* display a parse  tree

# *Words tokenization using NLTK*


* Text analysis relies heavily on tokenization. The text is divided into smaller bits called tokens during the tokenization process. The following methods can be used to tokenize the text: 
   * word – here, the user can find the most frequently used words in the text as well as determine which part of speech they belong to.
   * sentence - this form of tokenization enables the user to examine the entire statement. It demonstrates the relationship between words and their context.



```python
import nltk
```

```python
paragraphs = soup.find_all('p')
```

```python
for el in paragraphs:
    text = el.get_text().strip()
    if len(text) > 0:
        tokens = nltk.word_tokenize(text)
        tagged = nltk.pos_tag(tokens)
        print(tagged)
```

# *Analyzing the data*

```python
tagged = nltk.pos_tag(tokens)
```

```python
from collections import Counter
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize

dictionary_counter_num = Counter([j for i,j in tagged])
print(dictionary_counter_num)
```

# *Bar chart vizualization*

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.bar(range(len(dictionary_counter_num)), list(dictionary_counter_num.values()), align='center')
plt.xticks(range(len(dictionary_counter_num)), list(dictionary_counter_num.keys()))
plt.show()
```
